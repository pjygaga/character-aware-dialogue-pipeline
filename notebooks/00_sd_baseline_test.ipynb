```python
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    safety_checker=None
).to("cuda")

print("Model loaded!")
```

```python

base_appearance = "a young woman with bushy brown hair, brown eyes, wearing Hogwarts school uniform with Gryffindor tie, realistic style"

#neutral for test
prompt = f"{base_appearance}, neutral expression, looking forward"
```

```python
image = pipe(
    prompt,
    num_inference_steps=30,
    guidance_scale=7.5
).images[0]

image.save("sd_test_output.png")
image
```
